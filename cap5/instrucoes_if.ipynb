## Um exemplo simples

```python
# carros
carros = ['audi', 'bmw', 'ferrari', 'mercedes']

# exemplo de uso do for com if
for carro in carros:
    if carro == 'ferrari':
        print(carro.upper())
    else:
        print(carro.title())
```
